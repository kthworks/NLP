### Ksponspeech data 

Read sample data and listening 

```python
filepath = 'C:/Kosponspeech_train/KsponSpeech_01/KsponSpeech_0001/'

soundfile = filepath + 'KsponSpeech_000079.pcm'
textfile = filepath + 'KsponSpeech_000119.txt'

with open(soundfile, 'rb') as opened_pcm_file:
    buf = opened_pcm_file.read()
    pcm_data = np.frombuffer(buf, dtype = 'int16')
    wav_data = librosa.util.buf_to_float(pcm_data)
    
with open(textfile, 'r') as text_file:
    lines = text_file.readlines()

print(lines)    
plt.plot(wav_data)
ipd.Audio(wav_data, rate=16000)
```
